```python
#Run the below code in your notebook to check the installed version
import shap
import matplotlib.pyplot as plt
from pycaret.classification import *
from sklearn.utils import shuffle
import random
import pandas as pd
plt.rcParams['pdf.fonttype'] = 'truetype'
```

```python
sample_trials=list(range(3,101))
num_vars=[2,5,10,20,50,100,200,300,400]#,500,1000,2000,5000
```

```python
for x in sample_trials:

    df=pd.read_csv('./sampling_results/var10000_MD_valid_binary_normdata_3000cells_%d.csv' % (x),index_col=0)
    for n in num_vars:
        AUC = []
        Accuracy = []
        Recall = []
        Prec = []
        F1 = []
        Kappa = []
        MCC = []
        input_data=df.iloc[:,0:n]
        input_data['binary']=df['binary']
        exp1 = setup(input_data, target = 'binary',session_id=1, index=False)
        nb = create_model('nb')
        Xtest = get_config('X_test')
        ytest = get_config('y_test')
        predict_model(nb);
        AUC.append(pull()['AUC'][0])
        Accuracy.append(pull()['Accuracy'][0])
        Recall.append(pull()['Recall'][0])
        Prec.append(pull()['Prec.'][0])
        F1.append(pull()['F1'][0])
        Kappa.append(pull()['Kappa'][0])
        MCC.append(pull()['MCC'][0])
        AUC_df=pd.DataFrame(AUC)
        Accuracy_df=pd.DataFrame(Accuracy)
        Recall_df=pd.DataFrame(Recall)
        Prec_df=pd.DataFrame(Prec)
        F1_df=pd.DataFrame(F1)
        Kappa_df=pd.DataFrame(Kappa)
        MCC_df=pd.DataFrame(MCC)
        AUC_df.to_csv("./md_results/MD_var_model_results_%d_genes_AUC_%d.csv" % (n,x))
        Accuracy_df.to_csv("./md_results/MD_var_model_results_%d_genes_Acc_%d.csv" % (n,x))
        Recall_df.to_csv("./md_results/MD_var_model_results_%d_genes_Recall_%d.csv" % (n,x))
        Prec_df.to_csv("./md_results/MD_var_model_results_%d_genes_Prec_%d.csv" % (n,x))
        F1_df.to_csv("./md_results/MD_var_model_results_%d_genes_F1_%d.csv" % (n,x))
        Kappa_df.to_csv("./md_results/MD_var_model_results_%d_genes_Kappa_%d.csv" % (n,x))
        MCC_df.to_csv("./md_results/MD_var_model_results_%d_genes_MCC_%d.csv" % (n,x))
```
